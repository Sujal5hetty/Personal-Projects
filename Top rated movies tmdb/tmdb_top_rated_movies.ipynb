```python
import time



# Loop through 1 to 503 pages
for i in range(1, 504):
    # API request
    response = requests.get(
        f"https://api.themoviedb.org/3/movie/top_rated?api_key=ec2c952b86ae64b758130d466fe5001f&language=en&page={i}"
    )

    # Check if the response is successful
    if response.status_code == 200:
        data = response.json()

        # Check if 'results' exists in the response
        if "results" in data:
            # Create DataFrame from results
            topr_df = pd.DataFrame(data["results"])[
                ["id", "title", "release_date", "overview", "popularity", "vote_average", "vote_count"]
            ]

            # Concatenate with the main DataFrame
            df = pd.concat([df, topr_df], ignore_index=True)
        else:
            print(f"⚠️ No results found on page {i}")

    else:
        print(f"❌ Error on page {i}: {response.status_code}")

    # Add a small delay to prevent rate limiting
    time.sleep(0.5)
```
