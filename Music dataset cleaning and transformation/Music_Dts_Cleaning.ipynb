```python
import pandas as pd
data = {
'Song_ID': [101, 102, 103, 104, None, 106],
'Genre': ['Pop', 'Rock', 'Jazz', None, 'Hip-Hop', 'Rock'],
'Streams': ['1M', '500K', '750K', '1.2M', None, '900K']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Song_ID'] = df['Song_ID'].ffill() + df['Song_ID'].isna().cumsum()
df
```

```python
df.loc[5,"Song_ID"]=106.0
df
```

```python
df['Streams'].values
```

```python

def conv_to_v(value):
  d={'M':1000000,'K':1000}
  try:
    if isinstance(value,str) and value[-1] in d:
      res= float(value[:-1]) * d[value[-1]]
      return int(res)
  except ValueError:
    return None



df['Streams']= df['Streams'].apply(conv_to_v)
df
```

```python
df['Streams'] = df['Streams'].fillna(df['Streams'].mean())
df
```

```python
max_stream=df.loc[df['Streams'].nlargest(2).index[-1],'Genre']

df['Genre'] = df['Genre'].fillna(max_stream)
df
```

```python
df.to_csv("cleaned_music_data.csv")
```
